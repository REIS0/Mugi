# Sinewaves

```python
import soundfile as sf
import matplotlib.pyplot as plt
```

## 100hz

```python
data, samplerate = sf.read("files/100hz.flac")

plt.figure(figsize=(10,3))
plt.plot(data)
plt.show()
```

### Simple EA

```python
from src.algorithms.simple_ea import SimpleEA

simple_ea = SimpleEA(data, 50, fit=0.5, iterations=1000)

for i in range(10):
    generated = simple_ea.run()
    # blue is the generated, orange is the original
    plt.figure(figsize=(10,3))
    plt.plot(generated[1])
    plt.plot(data, alpha=0.7)
    plt.show()
```

### Euclidean EA

```python
from src.algorithms.euclidean_ea import EuclideanEA

euc_ea = EuclideanEA(data, 50, fit=10, iterations=1000)

# !! DEBUG
for _ in range(10):
    generated = euc_ea.run()
    print(generated[0])
    plt.figure(figsize=(10, 3))
    plt.plot(generated[1])
    plt.plot(data, alpha=0.5)
    plt.show()
```
